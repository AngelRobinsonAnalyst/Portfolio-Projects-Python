# 
What do you infer from this task? Can you infer an important parameter that the bank is using to approve loans?

```python
# Credit History seems to be the parameter that is being used to approve loans
# It appears that if an applicant has no credit history they are more likely to be denied
```

```python
approved = data[data.Loan_Status == "Y"]
approved_index = approved.index
approved_history = approved[approved.Credit_History == 1]
approved_history_index = approved_history.index
print("Total Applicants approved: \n",approved_index.size)
print(f"there are {approved_history_index.size} customers with credit history")

approved_nohistory = approved[approved.Credit_History == 0]
approved_nohistory_index = approved_nohistory.index
print(f"there are {approved_nohistory_index.size} customers with out credit history")
```

```python
denied = data[data.Loan_Status == "N"]
denied_index = denied.index
denied_history = denied[denied.Credit_History == 1]
denied_history_index = denied_history.index
print("Total Applicants denied: \n",denied_index.size)
print(f"there are {denied_history_index.size} customers with credit history")

denied_nohistory = denied[denied.Credit_History == 0]
denied_nohistory_index = denied_nohistory.index
print(f"there are {denied_nohistory_index.size} customers with out credit history")

#denied.sort_values(['ApplicantIncome'],ascending = False)
```
