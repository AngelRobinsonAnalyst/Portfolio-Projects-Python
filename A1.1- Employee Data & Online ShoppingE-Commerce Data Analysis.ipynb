```python
#IMPORTING DATA
import numpy as np
from numpy import genfromtxt
```

```python
shopping_data = genfromtxt("DS1_C4_S1_Shopping_Data_Challenge.csv",delimiter= ",",skip_header = 1)
shopping_data
#col[0] = customerID
#col[1] = age
#col[2] = annual income
#col[3] = spending score 
```

```python
#access customer ID
cus_id=shopping_data[:,0]
cus_id
```

```python
#access age
cus_age =shopping_data[:,1]
cus_age
```

```python
#access annual income
ann_inc = shopping_data[:,2]
ann_inc
```

```python
#access spending score
s_score = shopping_data[:,3]
s_score
```

```python
#TASK 1
#Display the details of the customers whose spending score is greater than 80.
above80_s_score = np.where (s_score > 80)
print(shopping_data[:,[0,3]][above80_s_score])
```

```python
#Display the total count of these customers.
num_above80_s_score = s_score[above80_s_score].size
print("There are",num_above80_s_score,"employees with a spending score above 80")
```

```python
#TASK 2
#Extract and display the data of the customers who are 20 to 24 years of age.
#NOTE: This data helps analyze age-related expenditures.
cus_age_group = np.where((cus_age >= 20)&(cus_age <25))
print(shopping_data[cus_age_group])
```

```python
#TASK 3
#Display the IDs of the customers with the lowest annual income and the customers with the highest.
lowest_ann_inc = min(ann_inc)
print("Lowest income: ",lowest_ann_inc)
highest_ann_inc =max(ann_inc)
print("Highest income:",highest_ann_inc)
```

```python
#TASK 3.1
#Display customer IDs who have the lowest salary
lowest_ann_incID = np.where(ann_inc == 15)
print(shopping_data[:,[0,2]][lowest_ann_incID])
```

```python
#TASK 3.2
#Display Customer ID who has the highest salary
highest_ann_incID = np.where(ann_inc == 137)
print(cus_id[highest_ann_incID])
print(shopping_data[:,[0,2]][highest_ann_incID])
```

```python
#TASK 4
#Display the details of the customers who are:
#CONDITIONS: less than 45 years of age
    #have an annual income greater than $35k 
    #have a spending score greater than 75

chosen_ones = np.where((cus_age < 45)&(ann_inc > 35)&(s_score > 75))
print(shopping_data[chosen_ones])
```

```python
#TASK 4.1
#How many such customers are there in the given data?
num_chosen_ones = cus_id[chosen_ones].size
print("There are",num_chosen_ones,"employees that meet these qualifications")
```

```python
#TASK 5
#During the Christmas season, the company wishes to offer random gift coupons valued between $10 and $50 to all its customers. 
gift_coupon = np.random.randint(10,50,size =(200,1) )
gift_coupon
print(gift_coupon.shape)
print(shopping_data.shape)
```

```python
#Add a new column for the gift coupons in the existing data
shopping_data = np.append(shopping_data,gift_coupon,axis = 1)
```
