```python
#TASK 1 IMPLEMENT NEGATIVE MARKING FOR INCORRECT ANSWERS
Q={"1-Number of Continents on Earth":"1:Seven 2:Four 3:Five 4:Six", 
   "2-Which country successfully landed a man on moon":"1:USA 2:Germany 3:France 4:England",
    "3-Which country is not present in the continent of North America":"1:Mexico 2:USA 3:Cannada 4:Brazil",
  "4-Which is the capital city of USA":"1:NYC 2:Washington DC 3:Austin 4:San Jose",
  "5-Which is the second most spoken language in USA":"1:English 2:Spanish 3:Mandarin 4:Italian",
   "6-Which is the American National Bird":"1:Parrot 2:Crane 3:Eagle 4:Swan",
  "7-Which of the following space agencies belongs to the USA":"1:BNSC 2:ISRO 3:DLR 4:NASA",
  "8-Which country shares the longest border with USA": "1:Mexico 2:Brazil 3:Cannada 4:Argentina",
  "9-Mt Rushmore has the sclupture of":"1:Former US Presidents 2:Hollywood stars 3:Military Personel 4:Scientists",
  "10-Which is the financial capital of USA": "1:SanFrancisco 2:Boston 3:Los Angeles 4:NYC"}
# The above dictionary Q contains the questions and their corresponding answer options.

'''Here are a few more questions and answers to consider adding to the above dictionary 

Which is the American National Bird (Parrot,Crane,Eagle,Swan) Ans: Eagle
Which country shares the longest border with USA (Mexico, Brazil, Cannada,Argentina) Ans: Cannada 
Which is the financial capital of USA (Sanfrancisco, Boston, Los Angeles, NYC) Ans : NYC
-Which of the following space agencies belongs to the USA (BNSC,ISRO,DLR,NASA) Ans : NASA
-Which is the second most spoken language in USA (English, Spanish, Mandarin, Italian) Ans : Spanish
-Which is the capital city of USA (NYC,Washington DC, Austin, San Jose) Ans : Washington DC
-Mt Rushmore has the sclupture of (Former US Presidents, Hollywood stars, Military Personel, Scientists) Ans: Former US Presidents

ANSWER KEY = 1:(1) 2:(1) 3:(4) 4:(2) 5:(2) 6:(3) 7:(4) 8:(3) 9:(1) 10:(4)

'''
A= list(Q.values()) # answer choices from Values in Q dictionary
CA = [1,1,4,2,2,3,4,3,1,4] # correct answers List referencing the answer number
score = 0 #SCORE COUNTER#
i = 0 # LOOP COUNTER #
myit = iter(A) # iterator to iterate over the answer choices
print("Hello! Welcome to my Quiz")
print("Lets Begin!")

for q in Q: # loop to display each question and answer choices    
    print("\n")    # SPACE #
    a=next(myit) # integrate answer choices into loop
    print("Question:",q) # Display question
    print(a) #Display: answer choices
    print("\n")    # SPACE  #
    Skip_Q = input("Do you want to skip this question? Enter (S) for YES or other to proceed: ") # display: ask to skip / User Input Section
    if (Skip_Q==str.lower("S")): # if user input = s or S #CONDITION 1#
            print("Ok, Here's the next question") # Display: OK next question
        
            print("\n") # SPACE #
            i = i + 1
            continue     # skip to the next question {next q in for loop}
            print(a) #Display: answer choices after condition 1 has passed
    elif (Skip_Q!=str.lower("S")): #if the above statement is false: proceed with program
            print("\n")
            
    ans = input("Okay! Lets answer the question! Enter the option between 1 and 4: ") # Display Enter the Option / User Input Section
    if(ans.isdigit()): # if user input = 1,2,3, or 4
            a = int(ans)  # type cast ans to an int so it can iterate if the statement is true
            
            if (a == CA[i]): # if user input = answer in answer choices {a==CA[i]}
                score = score + 1 #SCORE COUNTER# {add 1 to the score counter}
                print("That is correct!") #Display: that is correct                
            else: # if all the above statements are false{a!=CA[i]}, Display: incorrect answer
                print("Sorry, That is incorrect!")
                print("The correct answer is:",CA[i])
    else: # if the user input is not equal to 1,2,3, or 4 {if ans is not a digit}
            print("invalid entry")   # Dsiplay: entry is invalid
    i = i+1 #FOR LOOP COUNTER#
        
print("-----------------------------") # DISPLAYS AFTER LOOP IS COMPLETED

print("Your final score is: ",score,"out of 10 questions") # DISPLAYS FINAL SCORE ONCE LOOP IS COMPLETED
```
