# Task 1
Load the data into a DataFrame and check for missing values. 

```python
# Use pd.read_csv to read the csv file

car_data = pd.read_csv('DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')
car_data
```

# Task 2
Thomas, a potential customer from California, is interested in buying a low-priced car manufactured in 2014. 
- John needs to identify 100 lowest-priced cars from the year 2014 with CA as the state of registration.
- Help John create and display the details of potential cars.

```python
# create a variable to store the filtered dataframe
low_2014 = car_data[(car_data.year == 2014)&(car_data.state == 'ca')]

low_2014 
```

```python
# use sort_values to sort the data in ascending order
# use .head() to display the first 100 rows

low_2014 = low_2014.sort_values('mp',ascending = True).head(100)
low_2014
```

# Task 5
Thomas and several other customers seem to be interested in cars with black or gray colored interiors. 
- John wants to figure out if the cars with interiors in black or gray are priced higher or lower than cars having interiors of other colors.


### Determine the average price of all the 100 lowest-priced cars from the year 2014 based on their interior colors.

```python
# create 3 variables to store the car interior colors
car_black = low_2014[(low_2014.interior =='black')]
car_gray = low_2014[(low_2014.interior =='gray')]
car_other = low_2014[(low_2014.interior !='black')&(low_2014.interior !='gray')]

# create 3 variables to store the average prices based on the car interior colors
black_avg = car_black['mp'].mean()
gray_avg = car_gray['mp'].mean()
other_avg = car_other['mp'].mean()

# display the count of all cars per interior color along with their respective average prices
print('Total number of cars with black interiors: ',car_black.shape[0],"with an average price of: ",black_avg)
print('Total number of cars with gray interiors: ',car_gray.shape[0],"with an average price of: ",gray_avg)
print('Total number of cars with other colored interiors: ',car_other.shape[0],"with a combined average price of: ",other_avg)
```

```python
# FURTHER ANALYSIS
# create a variable to store the group of cars 
# use .groupby(by=)[] to group the cars by interior and order them based on the mp column 
# use the aggfunc .mean()
median_car = low_2014.groupby(by='interior')['mp'].median()

# used .sort_values() to sort the data in descending order
median_car.sort_values(ascending = False)
```

CONCLUSION
It seems that the black and white cars are the fourth and fifth most expensive out of the 6 interior color options. Their median prices are placed on the lower end of the data, which suggests that they are not priced higher than other car interiors.

# Task 6
Most buyers seem to prefer white and black colored cars. 
- John wants to analyze the color-wise availability of cars for each manufacturer from the years 2010 to 2015.

```python
# create a variable to store the black and white cars
# use the & and | operators

car_years = car_data[(car_data.year >= 2010)&(car_data.year <= 2015)&((car_data.color == 'black')|(car_data.color == 'white'))]

# sorted the data based on the year column in ascending order to show the oldest year first
car_years.sort_values('year',ascending = True)
```

```python
# create a variable to group the data
# use .groupby(by=)[] and the [] to group by 2 columns: make and color
# place the make in [] along with agg func .count() as the calculated field
car_years_group = car_years.groupby(by=['make','color'])['make'].count()
car_years_group
```

# Task 7
Eric, another customer, is interested in a good-condition BMW Sedan manufactured between 2010 and 2015. 
- John wants to display the details of all such cars that have a condition score >=4.5.

```python
# create a variable to store the bmw sedan cars 
car_bmw = car_data[
    (car_data.year >= 2010)&(car_data.year <= 2015)&(car_data.condition >= 4.5)&
    (car_data.make == 'BMW')&(car_data.body == 'Sedan')]
car_bmw
```

### Help John to figure out the year-wise maximum expected price (mp) of the BMW cars manufactured between the years 2010 and 2015 grouped by each model.

```python
# create a variable to store the grouped data
# use .groupby(by=)[] and [] to group by the year and model
# place the mp in [] along with agg func .mean() as the calculated field
car_bmw_group = car_bmw.groupby(by=['model','year'])['mp'].mean()
car_bmw_group

# sorted the data based on the mp column in ascending order to show the highest priced car first
car_bmw_group.sort_values(ascending = False)
```

John wants to recommend Eric to even consider SUVs from other manufacturers before finalizing the car.
- John needs to list the top 100 SUVs available between the years 2010 and 2015, based on their prices (in descending order) with an odometer reading <= 30000.

```python
car_data.dtypes
```

```python
# typecasting odometer from object to int

car_data = car_data.astype({'odometer': float})
car_data.dtypes
```

```python
# create a variable to store the suv cars
car_suv = car_data[
    (car_data.year >= 2010)&(car_data.year <= 2015)&
    (car_data.body == 'suv')&(car_data.odometer <= 30000.0)]
car_suv
```

```python
# use sort_values to sort the data in ascending order
# use .head() to display the first 100 rows

car_suv.sort_values('mp',ascending = False)
car_suv.head(100)
```
