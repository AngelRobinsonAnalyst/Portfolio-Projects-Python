### MODIFY THE 'Kerb_Weight' COLUMN

```python
# REMOVE KG FROM Kerb_Weight column USING str.replace() 
car_df['Kerb_Weight'] = car_df['Kerb_Weight'].str.replace(r"kg",'')
# CHANGE DTYPE TO INT USING .astype()
car_df = car_df.astype({'Kerb_Weight': int})
```

```python
# CONVERT WEIGHT TO POUNDS
car_df['Kerb_Weight'] = car_df['Kerb_Weight'] *2.2
car_df['Kerb_Weight']
```

### CLEANING THE POWER COLUMN

```python
# USE [] TO REFERENCE THE COLUMN FOLLOWED BY .str TO TRANSCRIBE THE VALUES INTO STRINGS,
    # NEXT USE THE .strip METHOD TO SEPERATE THE CHARACTERS AFTER THE DESIRED INTERGER
car_df['Power'] = car_df['Power'].str.strip()
# USE .replace METHOD TO REPLACE ANY CHARATER OR NUMBER  matches the regular expression r"[a-zA-Z]*@[0-9]*[a-zA-Z]*" with an empty string ''.
car_df['Power'] = car_df['Power'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'',regex = True)
```

```python
print("after cleaning")
car_df['Power']
```

```python
car_df = car_df.astype({'Power': float})
car_df.dtypes
```

### CLEANING THE TORQUE COLUMN

```python
car_df['Torque'] = car_df['Torque'].str.strip()
car_df['Torque'] = car_df['Torque'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'',regex = True)
car_df['Torque'] = car_df['Torque'].str.replace(r"-",'')
car_df['Torque'] = car_df['Torque'].str.replace(r".",'',regex= True)
car_df['Torque'] = car_df['Torque'].str.replace(r"rpm",'',regex= True)
car_df['Torque'] = car_df['Torque'].str.replace(r"RPM",'',regex= True)



car_df['Torque']
```

```python
car_df = car_df.astype({'Torque': float})
car_df['Torque']
```

# Task 3:
> In each column listed in Task 1, calculate and display the mean, median, standard deviation, and coefficient of variation for each manufacturer.


```python
# GROUP THE DATA BASED ON CAR MANUFACTURES
```

```python
# CREATE A VARIABLE TO STORE THE GROUPED DATA AND USE THE GROUP BY FUNCTION AND PASS THE COLUMN NAME AS A PARAMETER
grouped_make = car_df.groupby(['Make'])

make_names = car_df.Make.unique() # USE THE unique() FUNCTION TO EXTRACT THE UNIQUE VALUES OF THE MAKES FROM THE car_df DF
make_names
```

```python
# CREATE EMPTY 3  LISTS TO STORE CALCULATED COLUMNS AND ROW
cols = ['Power','Torque','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre']
row_head = ['Mean','Median','Standard Deviation','Coefficient of Variation']

# LOOP OVER THE DF TO CALCULATE THE CT OF EACH MANUFACTURER IN man_name DataFrame 
for name in make_names:
    make_data = pd.DataFrame()    #     CREATE A DATAFRAME man_data TO STORE THE CT MEASURES
    make_data = grouped_make.get_group(name) #     USE get_group() METHOD TO REASSIGN THE DATA TO A GROUPED DF GROUPED BY THE name

# CREATE EMPTY 3  LISTS TO STORE THE CENTRAL TENDENCIES
    Mean =[]
    Median =[]
    SD = []
    COV = []
# INITIALIZE AN IF STATEMENT TO ITERATE OVER A LIST IF THE LENGTH OF THE man_name df IS GREATER THAN 2
    if(len(make_data)>=2):
# ITERATE OVER THE LENGTH OF THE cols VARIABLE CONTAINING A LIST OF COLUMN HEADS
        for col in cols:
        # CREATE 3 VARIABLES TO CALCULATE THE CT AND APPEND THEM TO THE EMPTY LISTS ABOVE
            make_mean =make_data.loc[:,col].mean()
            Mean.append(make_mean)
            
            make_median = make_data.loc[:,col].median()
            Median.append(make_median)
            
            make_std = st.stdev(make_data.loc[:,col])
            SD.append(make_std)
            
            COV.append(make_std/make_mean*100)
        # CREATE A DATA FRAME man_ct_df TO STORE THE CT LISTS AND SET THE COLUMNS TO cols VARIABLE                            
        make_ct_df = pd.DataFrame([Mean, Median, SD, COV], columns = cols)
        make_ct_df.insert(0,'Measures', row_head, True) #    INSERT A NEW COLUMN Measures INTO THE row_heads VARIABLE CONTAINING THE CENTRAL TENDENCY LABELS

        
        print('Manufacturer: ',name, "\n")
        print(make_ct_df)
        print("************************************************************************************************")
```

> Identify at least three features that affect the mileage (using correlation analysis) of the cars of the following manufactures and visually represent them using the appropriate types of charts/graphs in Python:

- Tata

Illustrate with justification if these factors positively or negatively affect the mileage of a car.

### Mileage vs Displacement Scatterplot

```python
# DEFINE THE COLUMNS AND ROWS FOR THE FIGURE
make_names = ['Renault','Toyota','Tata'] # rows
data_cols = ['Cylinders','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque','Kerb_Weight']

# PLOT DIMENSIONS
fig, axes = plt.subplots(1,3, figsize=(10,2.5), dpi = 100, sharex=True, sharey=True ) # 1 row, 3 columns
data_colors = ['tab:orange','tab:blue','tab:green']
plt.suptitle('Mileage vs Displacement',y = 1.05, size = 16) # FIGURE TTILE, HEIGHT and width

# LOOP OVER THE make_names AND data_colors df TO FORMAT THE DATA INSIDE THE PLOT
    # ENUMERATE OVER A FLATTENED AXES CONTAINING THE make_names AND data_colors
for i, (ax, name, col) in enumerate(zip(axes.flatten(), make_names, data_colors)):
    make_data = grouped_make.get_group(name) 
    mpg = (make_data['City_Mileage_km_litre'] + make_data['Highway_Mileage_km_litre'])/2  # x axis
    make_corr = mpg.corr(make_data['Displacement']) # axis expression 
    # SHOW THE CORRELATION AMONG VARIABLES USING SCATTER PLOT
    ax.scatter(mpg,make_data['Displacement'], color=col)
    ax.set_title(name)
    ax.set_xlabel("R -" + str(make_corr)) # axis expression label
plt.tight_layout();
plt.show()
```

> Conclusion:
    The data shows a negative relationship between the Mileage and Displacement. 
    
    The mileage (x axis) increases as the Displacement (y axis) decreases.
    We can infer that the greater the Engine Displacement, the lower the MPG ratio.
